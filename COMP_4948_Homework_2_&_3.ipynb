## B. Driving Simulator Controls
The simulator allows users to drive around the track using keyboard presses as input. This behaves somewhat similar to a real vehicle with steering, throttle and brake inputs. During training testing, using the mouse controls to steer the vehicle resulted in better training data since it allowed for more continuous input instead of "on and off" such as a button press. This will be the method that we are going to use for gathering our training data.

Vehicle Controls:  
-- Drive Forward: UP arrow key  
-- Steer Left: LEFT arrow key or Left Mouse Click and Drag Left  
-- Steer Right: RIGHT arrow key or Left Mouse Click and Drag Left  
-- Brake or Reverse: DOWN arrow key

Record (Training Mode):  
-- R key
